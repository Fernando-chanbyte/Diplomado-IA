```python
import numpy as np
```

```python
x = np.random.rand(100)
y = np.random.rand(100)
```

```python
plt.scatter(x,y)
plt.title('Grafico de dispersion')
plt.show()
```

```python
datos = np.random.randn(10000)

plt.hist(datos,bins=100)
plt.xlabel('valor')
plt.ylabel('frecuencia')
plt.show()
```

```python
datos = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(datos,vert=True)
plt.show()
```

```python
import pandas as pd

data_range = pd.date_range(start='2023-01-01',end='2023-12-31',freq='D')
precios = np.cumsum(np.random.randn(len(data_range))) + 100
```

```python
data = pd.DataFrame({'Date':data_range,'precios': precios})
data.head()
```

```python
plt.plot(data['Date'],data['precios'], label='Precio')
plt.show()
```

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)

categorias = ['a', 'b', 'c', 'd']
valores = [5,7,3,8]

x2 = np.random.rand(50)
y2 = np.random.rand(50)


```

```python
fig, axs = plt.subplots(4,4, figsize = (15,10))

axs[0,0].plot(x,y1)
axs[0,0].set_title('Grafico de lineas')

axs[0,1].bar(categorias,valores)
axs[0,1].set_title('grafico de barras')

axs[1,0].plt.scatter(x2,y2)
axs[1,0].set.title('Grafico de dispersion')

axs[0,1].datos = [np.random.normal(0,std,100) for std in range(1,4)]
axs[0,1].set.boxplot(datos,vert=True)


plt.show()
```
